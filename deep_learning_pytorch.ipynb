```python
import torch.nn as nn
input_tensor = torch.tensor([[0.3471, 0.4547, -0.2356]])
linear_layer = nn.Linear(3, 2)
output_tensor = linear_layer(input_tensor)
print(output_tensor)
print(linear_layer.weight)
print(linear_layer.bias)
```

```python
import torch
import torch.nn as nn
input_tensor = torch.tensor([[7.0, 8.0, 3.0, 1.5, 8.9, 7.5, 6.0, 5.0]])
model = nn.Sequential(
    nn.Linear(8, 4),
    nn.Linear(4, 2)
)
output = model(input_tensor)
print(output)
```

```python
import torch
import torch.nn as nn
model = nn.Sequential(
    nn.Linear(9,4),
    nn.Linear(4,2),
    nn.Linear(2,1)
)
total = 0
for parameter in model.parameters():
  total += parameter.numel()
print(total)
```

```python
import torch
import torch.nn as nn

input_tensor = torch.tensor([[6]])
sigmoid = nn.Sigmoid()
output = sigmoid(input_tensor)

print(output)
```

```python
import torch
import torch.nn as nn

input_tensor = torch.tensor([[4.0, 0.0, 1.0]])
model = nn.Sequential(
    nn.Linear(3,4),
    nn.Linear(4,1),
    nn.Sigmoid()
)
output = model(input_tensor)

print(output)
```

```python
import torch
import torch.nn as nn

batch_size = 5
features = 4
classes = 3

tensor = torch.randn(batch_size, features, classes)
model = nn.Sequential(
    nn.Linear(classes, 2), # Corrected: in_features should be 'classes' (3)
    nn.Linear(2, classes),
    nn.Softmax(dim=-1)
)
output = model(tensor)
print(output)
```
